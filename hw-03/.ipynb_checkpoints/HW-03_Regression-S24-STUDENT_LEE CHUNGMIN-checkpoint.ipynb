# Part 3: One Variable Linear Regression (30 points)

In this part, we'll perform some one-variable linear regression analysis on the world happiness data.

&#9989;  **Question 3.1 (6 points)**: Using the `OLS` method in `statsmodels`, perform a linear regression to predict the `happiness_score` using the `healthy_life_expectancy` column separately for 2022 and 2023 and **display the results summaries** for both years. Remember that you may need to use the `add_constant()` method to make sure `OLS` fits a general line $y = ax+b$ to the data instead of a line through the origin $y = ax$. 

````python
# Put your code here

import statsmodels.api as sm

# For 2022
x_2022 = sm.add_constant(df_2022['healthy_life_expectancy'])
y_2022 = df_2022['happiness_score']
model_2022 = sm.OLS(y_2022, x_2022)
results_2022 = model_2022.fit()
print("Regression results for 2022:")
print(results_2022.summary())

# For 2023
x_2023 = sm.add_constant(df_2023['healthy_life_expectancy'])
y_2023 = df_2023['happiness_score']
model_2023 = sm.OLS(y_2023, x_2023)
results_2023 = model_2023.fit()
print("\nRegression results for 2023:")
print(results_2023.summary())
````

&#9989;  **Question 3.2 (6 points)**: Answer the following questions: 

1. What were the equations of the best fit lines for 2022 and 2023? 

2. Are the slopes and intercepts of the two fitted lines identical (between the 2 years)? Would you use the 2022 model to predict for 2023?

3. Based on the R-squared values for the two cases, how do you think the models fit the data?

<font size=+3>&#9998;</font> Put your answers here:

1. The equations of the best fit lines for 2022 and 2023, considering only countries in the Western Europe region, are as follows:
    * happiness_score_2022 = 2.8791 + 4.5626 * healthy_life_expectancy
    * happiness_score_2023 = 3.5502 + 5.4461 * healthy_life_expectancy


2. No, the slopes and intercepts of the two fitted lines are not identical between the years 2022 and 2023. For the year 2022, the intercept and the slope are 2.8791 and 4.5626, respectively. However, for the year 2023, they are 3.5502 and 5.4461, respectively. Since the slopes and intercepts are different, it implies that the relationships between happiness score and healthy life expectancy are different for the two years.
    * Therefore, using the 2022 model to predict for 2023 might not be appropriate due to these differences. Each year's model reflects the relationship observed in that specific year, and using a model from a different year to predict outcomes in another year may not accurately capture the changes or variations in the data. It's generally recommended to use the model built on data from the specific year you are interested in for making predictions within that year. Therefore, it would be more appropriate to use the 2023 model to predict happiness scores for 2023.


3. The R-squared value, also known as the coefficient of determination, measures the proportion of the variance in the dependent variable (happiness score) that is predictable from the independent variable (healthy life expectancy). A higher R-squared value indicates a better fit of the model to the data. For the year 2022 and 2023, they are 0.548, 0.558, respectively.
    * Both R-squared values are relatively close to each other, indicating that both models explain a moderate proportion of the variance in happiness scores based on healthy life expectancy. However, the R-squared value for the year 2023 (0.558) is slightly higher than that for the year 2022 (0.548), suggesting that the model for 2023 might have a slightly better fit to the data compared to the model for 2022.
    * Overall, both models seem to fit the data moderately well, but there may be other factors not included in the models that also influence happiness scores.

&#9989;  **Question 3.3 (6 points)**: Now make scatter plots of `happiness_score` (x-axis) vs. `healthy_life_expectancy` (on y-axis) for 2022 and 2023 on the same plot. Use different colors for the different years. Also plot the best fit lines on the same plot with same colors as the corresponding scatter plots. Label the axes, and add a legend, and give the plot a title.

````python
# Put your code here

import matplotlib.pyplot as plt

# Scatter plot for 2022
plt.scatter(df_2022['happiness_score'], df_2022['healthy_life_expectancy'], color='blue', label='2022')

# Scatter plot for 2023
plt.scatter(df_2023['happiness_score'], df_2023['healthy_life_expectancy'], color='red', label='2023')

# Plotting best fit lines
plt.plot(df_2022['happiness_score'], results_2022.predict(), color='blue', linestyle='--')
plt.plot(df_2023['happiness_score'], results_2023.predict(), color='red', linestyle='--')

# Labels and title
plt.xlabel('Happiness Score')
plt.ylabel('Healthy Life Expectancy')
plt.title('Happiness Score vs. Healthy Life Expectancy for 2022 and 2023')
plt.legend()
plt.grid()

plt.show()
````

&#9989;  **Question 3.4 (2 points)**: In which year are the happiness score predictions generally better?

<font size=+3>&#9998;</font> From the plot, it appears that the happiness score predictions are generally better for the year 2023. This inference is based on the fact that the regression line for 2023 seems to fit the data points more closely compared to the regression line for 2022. Additionally, the slightly higher R-squared value for 2023 also suggests a slightly better fit of the model to the data for that year.

**&#9989; Do this:** **Question 3.5 (4 points)**: Use `plot_regress_exog` to investigate the distribution of residuals in your model fits. Make sure to create a large enough figure so that everything is easily visible. Make separate plots for 2022 and 2023.

````python
# Put code here.

# For 2022
sm.graphics.plot_regress_exog(results_2022, 'healthy_life_expectancy', fig=plt.figure(figsize=(12, 8)))
plt.suptitle('Residuals Analysis for 2022', fontsize=16)
plt.show()

# For 2023
sm.graphics.plot_regress_exog(results_2023, 'healthy_life_expectancy', fig=plt.figure(figsize=(12, 8)))
plt.suptitle('Residuals Analysis for 2023', fontsize=16)
plt.show()
````

# Part 4: Multiple Regression (16 points)

In this part, we'll use multiple features to do predictions.

&#9989;  **Question 4.1 (5 points)**: Using the `OLS` method in `statsmodels`, perform a multivariable linear regression to predict the `happiness_score` based on `gdp_per_capita`, `healthy_life_expectancy`, `freedom_to_make_life_choices`, and `generosity`. Also, use the `add_constant()` method in `statsmodels` to ensure the model includes a constant term as well. Fit models to both the 2022 and 2023 data and display summaries of both results.

````python
# Put your code here

# For 2022
x_2022_mul = sm.add_constant(df_2022[['gdp_per_capita', 'healthy_life_expectancy', 'freedom_to_make_life_choices', 'generosity']])
y_2022_mul = df_2022['happiness_score']
model_2022_mul = sm.OLS(y_2022_mul, x_2022_mul)
results_2022_mul = model_2022_mul.fit()
print("Regression results for 2022 (Multivariable Linear Regression):")
print(results_2022_mul.summary())

# For 2023
x_2023_mul = sm.add_constant(df_2023[['gdp_per_capita', 'healthy_life_expectancy', 'freedom_to_make_life_choices', 'generosity']])
y_2023_mul = df_2023['happiness_score']
model_2023_mul = sm.OLS(y_2023_mul, x_2023_mul)
results_2023_mul = model_2023_mul.fit()
print("\nRegression results for 2023 (Multivariable Linear Regression):")
print(results_2023_mul.summary())
````

&#9989;  **Question 4.3 (2 points)**: Perform the multivariable linear regression for 2023, but this time with only the **most significant features** (p-values less than 0.05). Display the summary of these results.

````python
# Put your code here

# Selecting only the significant features for 2023
x_2023_significant = sm.add_constant(df_2023[['gdp_per_capita', 'healthy_life_expectancy', 'freedom_to_make_life_choices']])
y_2023_significant = df_2023['happiness_score']

# Fitting the model
model_2023_significant = sm.OLS(y_2023_significant, x_2023_significant)
results_2023_significant = model_2023_significant.fit()

# Displaying the summary of results
print("Regression results for 2023 with significant features only:")
print(results_2023_significant.summary())
````

Overall, while the reduced model may sacrifice a small amount of predictive power compared to the original model, its simplicity and potential for improved stability make it a valuable alternative, especially in cases where interpretability and computational efficiency are priorities.

&#9989;  **Question 4.5 (3 points)**: Create three `.graphics.plot_regress_exog` figures, one for each of the features (columns of original dataframe) in your reduced model. Pay attention to the top two plots: the fitted values figure and the residual plot. 

````python
# Put your code here.

import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.regressionplots import plot_regress_exog

# Function to plot regression figures
def plot_regression_figures(feature_name, result):
    fig = plt.figure(figsize=(12, 6))
    plot_regress_exog(result, feature_name, fig=fig)

# Plot figures for each feature
features = ['gdp_per_capita', 'healthy_life_expectancy', 'freedom_to_make_life_choices']
for feature in features:
    plot_regression_figures(feature, results_2023_significant)  # Using the results of the reduced model

plt.show()
````

# Part 5: Logistic Regression (17 points)

In this part, we'd like to use logistic regression to classify whether a person has diabetes or not. Logistic regression (as we've learned so far in class) does binary classification. 



&#9989;  **Question 5.1 (2 points)**: We will work with data that is available at https://www.kaggle.com/datasets/aemyjutt/diabetesdataanslysis?select=diabetes.csv. 

**You'll need to download the following file**:

`https://raw.githubusercontent.com/ravisha3/CMSE202S24Homework3/main/diabetes.csv`

**Do This**: In the cell below, type the code for downloading the data from inside your notebook and also code for loading the data into a pandas dataframe.

````python
# Put your code here
!curl -O https://raw.githubusercontent.com/ravisha3/CMSE202S24Homework3/main/diabetes.csv

# Loading the data into a pandas dataframe
df_diabetes = pd.read_csv("diabetes.csv", delimiter=',')
df_diabetes.head()
````

&#9989;  **Question 5.2 (3 points)**:

Create a `Pandas` `Series` called `labels` which has data from the `Outcome` column of the DataFrame. Also, create a `Pandas` `DataFrame` called `features` which consists of just the columns `Glucose`, `BloodPressure`, `BMI`, and `Age`. Display the labels and features to make sure you did this correctly.

````python
# Put your code here.

# Creating a Pandas Series called labels from the Outcome column of the DataFrame
labels = df_diabetes['Outcome']

# Creating a Pandas DataFrame called features with selected columns
features = df_diabetes[['Glucose', 'BloodPressure', 'BMI', 'Age']]

# Displaying the labels
print("Labels:")
print(labels.head())

# Displaying the features
print("\nFeatures:")
print(features.head())
````

&#9989;  **Question 5.3 (4 points)**: Split your data into a training and testing set with a training set representing 80% of your data. For reproducibility, set the `random_state` argument to `0`. Print the shapes of the training features, the testing features, the training labels, and the testing labels to show you have the right number of entries in each of the four variables.

````python
# Put your code here

from sklearn.model_selection import train_test_split

# Splitting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=0)

# Printing the shapes of training features, testing features, training labels, and testing labels
print("Shape of training features:", x_train.shape)
print("Shape of testing features:", x_test.shape)
print("Shape of training labels:", y_train.shape)
print("Shape of testing labels:", y_test.shape)
````

&#9989;  **Question 5.4 (4 points)**: Now, train a logistic regression model using your training features and training labels. **Display the summary**.

````python
# Put your code here

from sklearn.linear_model import LogisticRegression

# Creating and training a logistic regression model
logistic_model = LogisticRegression()
logistic_model.fit(x_train, y_train)

# Displaying the summary
print("Logistic Regression Model Summary:")
print("Coefficients:", logistic_model.coef_)
print("Intercept:", logistic_model.intercept_)
````

&#9989;  **Question 5.5 (4 points)**: Finally, test your logistic regression model using your testing features and testing labels. Display the fraction of testing data points that were correctly predicted. 

````python
# Put your code here

# Predicting the labels for testing features
y_pred = logistic_model.predict(x_test)

# Calculating the fraction of testing data points that were correctly predicted
accuracy = (y_pred == y_test).mean()
print("Fraction of testing data points correctly predicted:", "with the accuracy of", accuracy)
````
